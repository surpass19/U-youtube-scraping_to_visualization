# DataFrameで保存

```python
output = []

for num, item in enumerate(item_list):
    item_url = item.find(class_="p-item_visual is-biggerlinkBigger s-biggerlinkHover_alpha").get("href")

    item_title = item.find(class_="p-item_visual is-biggerlinkBigger s-biggerlinkHover_alpha").find("img").get("alt")
    
    output.append([item_url,item_title])
    
pd.DataFrame(output, columns = ["item_url", "item_title"])
```

### 辞書使ってもできた気がする...こっちのが楽そう

```python
output = []

for num, item in enumerate(item_list):
    item_dic = {}
    item_url = item.find(class_="p-item_visual is-biggerlinkBigger s-biggerlinkHover_alpha").get("href")
    item_dic["item_url"] = item_url
    item_title = item.find(class_="p-item_visual is-biggerlinkBigger s-biggerlinkHover_alpha").find("img").get("alt")
    item_dic["item_title"] = item_title
    
    item_picture = item.find(class_="p-item_visual is-biggerlinkBigger s-biggerlinkHover_alpha").find("img").get("data-src")
    item_dic["item_picture"] = item_picture
    
    item_maker = item.find(class_="p-item_maker").text
    item_dic["item_maker"] = item_maker

    item_price = item.find(class_="p-item_price").text
    item_dic["item_price"] = item_price
    
    output.append(item_dic)
    
pd.DataFrame(output)
```

### 出てきた, もっと楽な方法

```python
output = []

for num, item in enumerate(item_list):
    item_url = item.find(class_="p-item_visual is-biggerlinkBigger s-biggerlinkHover_alpha").get("href")
    item_title = item.find(class_="p-item_visual is-biggerlinkBigger s-biggerlinkHover_alpha").find("img").get("alt")
    
    item_picture = item.find(class_="p-item_visual is-biggerlinkBigger s-biggerlinkHover_alpha").find("img").get("data-src")
    
    item_maker = item.find(class_="p-item_maker").text

    item_price = item.find(class_="p-item_price").text
    
    output.append({
        "item_url":item_url,
        "item_title":item_title,
        "item_picture":item_picture,
        "item_maker":item_maker,
        "item_price":item_price,
    })
    
pd.DataFrame(output)
```

```python
def next_url(res_bs):
    url = res_bs.find(class_="p-pager_btn p-pager_btn_next").find("a").get("href")
    return url
```

# URLが終わるまで繰り返す

```python
import time
```

```python
#価格.com
url = "https://kakaku.com/search_results/laptop/?category=0001%2C0004"
```

```python
for num in range(1, 50):
    print(f"{num}回目")
    res = requests.get(url)
    res_bs = bs(res.content, "html.parser")
    time.sleep(5)
    url = next_url(res_bs)
    if url is None:
        break
```
